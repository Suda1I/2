```python
import numpy as np

# 1. Задаем параметры
R = [0, 10, 20, 10, 15, 20, 10] # R[1]...R[6]
E = [0, 10, 20, 30]             # E[1]...E[3]

# 2. Формируем матрицу коэффициентов A (по вашему изображению)
# Порядок переменных в системе: I1, I2, I3, I4, I5, I6
A = np.array([
    [1,  0,  0, -1, -1,  0],  # -I5 - I4 + I1 = 0
    [0, -1,  0,  1,  0,  1],  # I4 - I2 + I6 = 0
    [0,  1, -1,  0,  1,  0],  # I2 - I3 + I5 = 0
    [0, -R[4], 0, -R[2], R[1], 0], # I5*R1 - I4*R2 - I2*R4 = -E3 - E1
    [R[6], 0,  0,  R[2], 0, -R[3]], # I4*R2 + I1*R6 - I6*R3 = -E2
    [0, -R[4], -R[5], 0, 0, -R[3]]  # -I6*R3 - I3*R5 - I2*R4 = -E2 - E1
], dtype=float)

# 3. Вектор свободных членов B
B = np.array([0, 0, 0, -E[3]-E[1], -E[2], -E[2]-E[1]], dtype=float)

def solve_cramer(A, B):
    det_A = np.linalg.det(A)
    if np.isclose(det_A, 0):
        return "Система не имеет единственного решения"

    results = []
    for i in range(len(B)):
        Ai = A.copy()
        Ai[:, i] = B # Заменяем i-й столбец вектором B
        results.append(np.linalg.det(Ai) / det_A)
    return np.array(results)

# Решаем
currents = solve_cramer(A, B)

print("Токи в ветвях (I1 - I6):")
for i, val in enumerate(currents, 1):
    print(f"I{i} = {val:.4f} A")

# Проверка через встроенную функцию
check = np.linalg.solve(A, B)
print("\nПроверка (linalg.solve):", np.allclose(currents, check))
```
